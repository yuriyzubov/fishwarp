```python
import zarr
import os
import sklearn
import numpy as np
from metrics import ncc, mutual_information, metrics_sitk
```

```python
# load reference sample + before and after registration data
sample_num = 0
prefix = "path/to/project/data"

ref = zarr.open(os.path.join(prefix, f'src/test_samples/sample{sample_num}', 'confocal.zarr', 's0'), 'r')[:]
no_deform = np.float32(zarr.open(os.path.join(prefix, f'src/test_samples/sample{sample_num}', 'anatomy.zarr', 's0'), 'r')[:])
deform = zarr.open(os.path.join(prefix, f'registration/test_samples/sample{sample_num}/anatomy2confocal', 'deformed.zarr'), 'r')[:]
```

```python
# metrics
metrics = [
    ('Normalized Cross-Correlation', lambda x: ncc(ref, x)), 
    ('Mutual Inoformation', lambda x: mutual_information(ref, x)),
    ('ITK metrics', lambda x: metrics_sitk(ref, x))
]
```

```python
# check if registration improved alignment of confocal and functional anatomy volumes
samples = [
        deform,
        no_deform
        ]

for (method_name, metric) in metrics:
    results = list(map(metric, samples))
    print(f"{method_name}: {results}")
```
